```python
import random

colors = {
    "Black": '\x1b[90m',
    "Blue": '\x1b[94m',
    "Cyan": '\x1b[96m',
    "Green": '\x1b[92m',
    "Magenta": '\x1b[95m',
    "Red": '\x1b[91m',
    "White": '\x1b[97m',
    "Yellow": '\x1b[93m',

}


magic_cards = [
    {"name": "Up Attack", "ability": "increase_attack"},
    {"name": "Up defense", "ability": "increase_defense"},
    {"name": "Invicible", "ability": "invincibility"},
    {"name": "Skip Turn", "ability": "skip_turn"},
    {"name": "Poison", "ability": "poison"}
]


class Part:

    def __init__(self, name, attack_level, defense_level,energy_consumption):
        self.name = name
        self.attack_level = attack_level
        self.defense_level = defense_level
        self.energy_consumption = energy_consumption #cantidad de energia utilizada
        self.available = True

    def get_status_dict(self):
      formatted_name = self.name.replace(" ","_").lower()
      return {
          "{}_name".format(formatted_name): self.name.upper(),
          "{}_status".format(formatted_name): self.is_available(),
          "{}_attack".format(formatted_name): self.attack_level,
          "{}_defense".format(formatted_name): self.defense_level,
          "{}_energy_consump".format(formatted_name): self.energy_consumption,
          }


    def reduce_edefense(self):

        if self.defense_level <= 30:
          print(f"Warning! The defense level of {self.name} is very low: {self.defense_level} %")

        else:
          print(f"Defense level of {self.name} is stable")


    def is_available(self):
      return not self.defense_level <= 0

class Robot:

    def __init__(self, name, color_code):
        self.name = name
        self.color_code = color_code
        self.energy = 100
        self.parts = [
            Part("Head", attack_level=5, defense_level=10, energy_consumption=5),
            Part("Weapon", attack_level=15, defense_level=30, energy_consumption=10),
            Part("Left Arm", attack_level=3, defense_level=21, energy_consumption=10),
            Part("Right Arm", attack_level=6, defense_level=20, energy_consumption=10),
            Part("Left Leg", attack_level=4, defense_level=20, energy_consumption=15),
            Part("Right Leg", attack_level=8, defense_level=20, energy_consumption=15),

        ]

    def greet(self):
        print("Hello my name is:", self.name)

    def print_energy(self):
      print(self.energy)


    def attack(self, enemy_robot, part_to_use, part_to_attack):
      enemy_robot.parts[part_to_attack].defense_level -=  self.parts[part_to_use].attack_level
      self.energy -= self.parts[part_to_use].energy_consumption

    def is_on(self):
      return self.energy >= 0


    def is_there_available_parts(self):
      for part in self.parts:
        if part.is_available():
          return True
      return False

    def print_status(self):
      print(self.color_code)
      str_robot = robot_art.format(**self.get_part_status())
      self.greet()
      self.print_energy()
      print(str_robot)
      print(colors["White"])


    def get_part_status(self):
      part_status = {}
      for part in self.parts:
        status_dict = part.get_status_dict()
        part_status.update(status_dict)
      return part_status

class Player:
  def __init__(self, name, robot):
    self.robot = robot
    self.name = name
    self.magic_cards = []
    self.invincible = False  # Nuevo atributo para rastrear la invencibilidad


  def  receive_magic_cards(self):
    # Recibo 2 cartas aleatorias al inicio del juego
    self.magic_cards = random.sample(magic_cards, 2)

  def use_magic_card(self, enemy_player):
    # el jugador elige una para usar
        if self.magic_cards:
            print(f"{self.name}, it's your turn. Your available cards are:")
            for i, card in enumerate(self.magic_cards):
                print(f"{i + 1}. {card['name']}")
            while True:
                choice = input("Choose the number of the card you want to use (0 to skip): ")
                if choice == '0':
                    print("You skipped using cards in this turn.")
                    break
                elif choice.isdigit():
                    choice = int(choice)
                    if 1 <= choice <= len(self.magic_cards):
                        card = self.magic_cards.pop(choice - 1)
                        self.apply_magic_card(card, enemy_player)
                        break
                print("Invalid option. Try again.")
        else:
            print("No cards available for this turn.")


  def apply_magic_card(self, card, enemy_player):
     #habilidades de carta
        if card["ability"] == "increase_attack":
            self.increase_attack()
        elif card["ability"] == "increase_defense":
            self.increase_defense()
        elif card["ability"] == "invincibility":
            self.apply_invincibility()
        elif card["ability"] == "skip_turn":
            self.skip_turn(enemy_player)
        elif card["ability"] == "poison":
            self.poison(enemy_player)


  def apply_invincibility(self):
        self.invincible = True
        print(f"{self.name} is invincible for one turn.")

  def increase_attack(self):
    print(f"{self.name} has increased the attack of a part by 50%.")
    while True:
        print("Which part's attack do you want to increase?")
        part_to_increase = input("Choose a number for the part: ")
        part_to_increase = int(part_to_increase)
        if part_to_increase not in range(len(self.robot.parts)):
            print("Invalid option. Please choose a valid part number.")
        else:
            current_part = self.robot.parts[part_to_increase]
            current_part.attack_level += current_part.attack_level * 0.5
            print(f"{current_part.name} attack increased to {current_part.attack_level}")
            break

  def increase_defense(self):
        print(f"{self.name} you increased your defense in a part 60%.")
        while True:
            print("Which part's defense do you want to increase?")
            part_to_increase = input("Choose a number for the part: ")
            part_to_increase = int(part_to_increase)
            if part_to_increase not in range(len(self.robot.parts)):
                print("Invalid option. Please choose a valid part number.")
            else:
                current_part = self.robot.parts[part_to_increase]
                current_part.defense_level += current_part.defense_level * 0.6
                print(f"{current_part.name} defense increased to {current_part.defense_level}")
                break


  def apply_invincibility(self):
        print(f"{self.name} is invincible for one turn.")

  def skip_turn(self, enemy_player):
        print(f"{self.name} has jumped the turn of {enemy_player.name}.")

  def poison(self, enemy_player):
        print(f"{self.name} ha envenenado a {enemy_player.name}.")

def play():
    playing = True
    print("Welcome to the game")

    robot_one = Robot("Jarvis", colors["Red"])
    robot_two = Robot("Friday", colors["Yellow"])

    player_one = Player(robot_one.name, robot_one)
    player_two = Player(robot_two.name, robot_two)

    player_one.receive_magic_cards()
    player_two.receive_magic_cards()

    round = 0

    while playing:

        if round % 2 == 0:
            current_player = player_one
            current_robot = robot_one
            enemy_robot = robot_two
        else:
            current_player = player_two
            current_robot = robot_two
            enemy_robot = robot_one


        current_robot.print_status()
        current_player.use_magic_card(current_robot)

        if current_player.invincible:
            print(f"{current_player.name} is currently invincible.")
            current_player.invincible = False  # Desactivar invincibilidad después del turno



        while True:
            print("What part shall I use to attack?")
            part_to_use = input("Choose a number part: ")
            part_to_use = int(part_to_use)
            if part_to_use not in range(len(current_robot.parts)):
                print("Invalid option. Please choose a valid part number (0,1,2,3,4,5).")
            else:
                break

        enemy_robot.print_status()

        while True:
            print("Which part of the enemy shall we attack?")
            part_to_attack = input("Choose an enemy part to attack: ")
            part_to_attack = int(part_to_attack)
            if part_to_attack not in [0, 1, 2, 3, 4, 5]:
                print("Invalid option. Please choose a valid part number (0,1,2,3,4,5).")
            else:
                break
        current_robot.attack(enemy_robot, part_to_use, part_to_attack)
        round += 1

        if not robot_one.is_on() or not robot_two.is_on():
            playing = False
            winner = player_one if robot_two.is_on() else player_two
            print("Congratulations, you won!")
            print(winner.name)
play()

```
